## Global Human Trafficking - Time Series Analysis

Analysis different features in relation to the 'yearofRegistration'

```python
import numpy as np
import pandas as pd
import seaborn as sns
import plotly
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
from pycountry_convert import country_alpha2_to_country_name, country_name_to_country_alpha3
import os
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("human_trafficking.csv")
data.head()
```

```python
data.replace('-99', np.nan, inplace=True)
data.replace(-99, np.nan, inplace=True)
```

```python
def get_alpha3(col):
    try:
        iso_3 =  country_name_to_country_alpha3(col)
    except:
        iso_3 = 'Unknown'
    return iso_3

def get_name(col):
    try:
        name =  country_alpha2_to_country_name(col)
    except:
        name = 'Unknown'
    return name
```

```python
data['country'] = data['citizenship'].apply(lambda x: get_name(x))
data['alpha_3'] = data['country'].apply(lambda x: get_alpha3(x))
```

```python
counts = data['yearOfRegistration'].value_counts()
sorted_counts = {}
for key in sorted(counts.keys()):
    sorted_counts.update({key:counts[key]})
    
plt.plot(sorted_counts.keys(), sorted_counts.values())
plt.title('Number of cases by Year')
plt.xlabel('Year')
plt.ylabel('Number of Registered cases')
plt.show()
```

According to the dataset, the number of registered cases reached a peak in the year 2016. 

```python
cm = sns.light_palette("red", as_cmap=True)
table = pd.pivot_table(data, values='Datasource', index='country',
                    columns='yearOfRegistration', aggfunc='count', fill_value=0)
table.style.background_gradient(cmap=cm)
```

```python
data['countryexp'] = data['CountryOfExploitation'].apply(lambda x: get_name(x))
data['alpha_3'] = data['countryexp'].apply(lambda x: get_alpha3(x))
```

```python
cm = sns.light_palette("red", as_cmap=True)
table = pd.pivot_table(data, values='Datasource', index='countryexp',
                    columns='yearOfRegistration', aggfunc='count', fill_value=0)
table.style.background_gradient(cmap=cm)
```
